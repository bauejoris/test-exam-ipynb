The following Python code creates a Pandas DataFrame `df` that contains 100 datapoints (`P1--P100`) with each 10 attributes (`F0--F9`) and a binary class label `label`.

The following code splits the dataset `df` into a train and test set at random.

The features for the train set are in `X_train` with the label in `y_train`.
The features for the test set are in `X_test` with the label in `y_test`.

````python
from sklearn.model_selection import train_test_split

X = df
y = X.pop("label")

X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.5, random_state=42)
````

**question 11 (4pts)**

We want to train a logistic regression model on the train set and evaluate the prediciton performance of the fitted model on the test set. Standardize the features, train the model and print the accuracy of the test set predictions. Use Scikit-learn.

````python
#start code here
#standardize (put them in the same scale)
from sklearn.preprocessing import StandardScaler
scaler_std = StandardScaler()
scaler_std.fit(X_train)
X_train_standardize = scaler_std.transform(X_train)
X_train_standardize = pd.DataFrame(X_train_standardize, columns=X_train.columns)
X_test_standardize = pd.DataFrame(scaler_std.transform(X_test),columns=X_test.columns)
#train
from sklearn.linear_model import LogisticRegression
cls_std = LogisticRegression()
cls_std.fit(X_train_standardize,y_train)
#accuracy
print("Accuracy: {}".format(cls_std.score(X_test_standardize,y_test)))
#end code here
````

**question 12 (1pt)**

Print the test set class predictions of the model.

````python
#start code here
predictions_standardize = cls_std.predict_proba(X_test_standardize)
print("class predictions: {}".format(predictions_standardize))
#end code here
````
